# Import

```python
import numpy as np  # Gestion des tableaux  
import matplotlib   # Gestion de l'affichage des graphiques
import matplotlib.pyplot as plt  # Gestion de l'affichage des graphiques
import pandas as pd # Gestion des dataframe pour un joli affichage
import seaborn as sns # Gestion des graphiques pour joli affichage
```

```python
from sklearn.datasets import load_breast_cancer
```

# Préparation de la base de données - assignation des étiquettes

```python
cancer["data"].shape

X = cancer["data"]
y = cancer["target"]

print("X.shape :", X.shape)
print("Y.shape :", y.shape)
```

```python
X_df = pd.DataFrame(X)
y_df = pd.DataFrame(y)

print(X_df.head(3))
print(y_df.head(3))

print(X_df.describe())
print(y_df.describe())
```

```python
print(X_df.dtypes)
print(y_df.dtypes)
```

# Prépatration des ensembles d'apprentissage et de test

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X,y) #test_size = 0.25
```

# Pre-processing

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

scaler.fit(X_train)

X_train = scaler.transform(X_train)
X_test = scaler.transform(X_test)
```

```python
X_train_df = pd.DataFrame(X_train)

print(X_train_df.describe())
```

```python
X_test_df = pd.DataFrame(X_test)

print(X_test_df.describe())
```

# Apprentissage

```python
from sklearn.metrics import f1_score, accuracy_score, roc_auc_score
```

## Premier apprentissage

### Entrainement

```python
from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes = (30,30,30))

mlp.fit(X_train, y_train)

# Attention : message d'avertissement :
# le nombre d'itération rentré ne permet pas de minimiser la fonction de
# coût. Cela pourrait affecter la pertinence du modèle. Essayer avec
# plus d'itérations ( > 200) ?
```

### Evaluation de la performance sur l'ensemble de test

```python
predictions = mlp.predict(X_test)

from sklearn.metrics import classification_report, confusion_matrix

cm = confusion_matrix(y_test, predictions)

plt.figure(figsize=(4, 3))
sns.heatmap(cm, annot=True, cmap="Reds")
plt.xlabel("Prédictions")
plt.ylabel("Valeurs réelles")
plt.title("Matrice de Confusion")
plt.show()

# à la vue de la matrice, on a une bonne sensibilité (précision): tous es patients
# atteints d'une tumeurs ont été détecté. Néanmoins, effort à faire sur
# la spécificité (rappel): des patients non atteints de tumeurs ont été classés
# comme ayant une tumeurs.
```

```python
print(classification_report(y_test, predictions))
```

### Tentative d'amélioration de la performance

#### Exploration nombres de couches pour 30 neuronnes/couches (nombre de neuronnes/couches arbitraire)

```python
couches = np.arange(1, 101, 4)

perf_train = []
perf_test = []

for i in couches :
    mlp_temp = MLPClassifier(hidden_layer_sizes=(30,) * i, random_state = 0)
    mlp_temp.fit(X_train, y_train)
    
    train_f1score = f1_score(mlp_temp.predict(X_train), y_train)
    test_f1score = f1_score(mlp_temp.predict(X_test), y_test)
    
    perf_train.append(train_f1score)
    perf_test.append(test_f1score)
    

plt.figure(figsize=(10, 6))
plt.plot(couches, perf_train, label='F1 Score sur l\'ensemble d\'entraînement')
plt.plot(couches, perf_test, label='F1 Score sur l\'ensemble de test')
plt.xlabel('Nombre de couches cachées')
plt.ylabel('Score F1')
plt.title('Score F1 en fonction du nombre de couches cachées (30 neu./cou.)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
index_meilleur_perf_test = np.argmax(perf_test)  # Indice du score F1 maximal pour l'ensemble de test
meilleur_couche = couches[index_meilleur_perf_test]  # Nombre de couches correspondant
meilleur_score = perf_test[index_meilleur_perf_test]  # Score F1 maximal

print(f"La meilleure configuration est de {meilleur_couche} couches cachées avec un score F1 de {meilleur_score:.4f}.")
```

#### Exploration du nombre de neuronnes pour meilleurs nombres de couches

```python
neuronnes = np.arange(2, 103, 4) # variation du nombres de neuronnes
                                 # entre 2 et 102 par 4
perf_train = []
perf_test = []

for i in neuronnes :
    mlp_temp = MLPClassifier(hidden_layer_sizes=(i,) * meilleur_couche, random_state = 0)
    mlp_temp.fit(X_train, y_train)
    
    train_f1score = f1_score(mlp_temp.predict(X_train), y_train)
    test_f1score = f1_score(mlp_temp.predict(X_test), y_test)
    
    perf_train.append(train_f1score)
    perf_test.append(test_f1score)
    

plt.figure(figsize=(10, 6))
plt.plot(neuronnes, perf_train, label='F1 Score sur l\'ensemble d\'entraînement')
plt.plot(neuronnes, perf_test, label='F1 Score sur l\'ensemble de test')
plt.xlabel('Nombre de neuronnes par couche')
plt.ylabel('Score F1')
plt.title('Score F1 en fonction du nombre de neuronnes (9 couches cachées)')
plt.legend()
plt.grid(True)
plt.show()
```

```python
index_meilleur_perf_test = np.argmax(perf_test)  # Indice du score F1 maximal pour l'ensemble de test
meilleur_nb_neuronnes = neuronnes[index_meilleur_perf_test]  # Nombre de couches correspondant
meilleur_score = perf_test[index_meilleur_perf_test]  # Score F1 maximal

print(f"La meilleure configuration est de {meilleur_nb_neuronnes} neuronnes avec un score F1 de {meilleur_score:.4f}.")
```

#### Exploration du nombre d'itérations pour meilleurs nombres de couches et de neuronnes

```python
iterations = np.arange(50, 500, 10) # variation du nombres de neuronnes
                                 # entre 2 et 102 par 4
perf_train = []
perf_test = []

for i in iterations :
    mlp_temp = MLPClassifier(hidden_layer_sizes=(meilleur_nb_neuronnes,) * meilleur_couche, 
                             max_iter = i, 
                             random_state = 0)
    mlp_temp.fit(X_train, y_train)
    
    train_f1score = f1_score(mlp_temp.predict(X_train), y_train)
    test_f1score = f1_score(mlp_temp.predict(X_test), y_test)
    
    perf_train.append(train_f1score)
    perf_test.append(test_f1score)
    

plt.figure(figsize=(10, 6))
plt.plot(iterations, perf_train, label='F1 Score sur l\'ensemble d\'entraînement')
plt.plot(iterations, perf_test, label='F1 Score sur l\'ensemble de test')
plt.xlabel("Nombre d'itérations")
plt.ylabel('Score F1')
plt.title("Score F1 en fonction du nombre d'itérations (9 couches cachées, 14 neu.cou.)")
plt.legend()
plt.grid(True)
plt.show()
```

```python
index_meilleur_perf_test = np.argmax(perf_test)  # Indice du score F1 maximal pour l'ensemble de test
meilleur_nb_iteration = iterations[index_meilleur_perf_test]  # Nombre de couches correspondant
meilleur_score = perf_test[index_meilleur_perf_test]  # Score F1 maximal

print(f"La meilleure configuration est de {meilleur_nb_iteration} itérations avec un score F1 de {meilleur_score:.4f}.")
```

### Usage de la validation croisée

```python
from sklearn.model_selection import cross_val_score
```

```python
mlp_meilleur = MLPClassifier(hidden_layer_sizes=(meilleur_nb_neuronnes,) * meilleur_couche, 
                             max_iter = meilleur_nb_iteration, 
                             random_state = 0)

cross_val_score(mlp_meilleur, X_train, y_train, cv = 3, scoring = "accuracy").mean()
```

#### Meilleur nombre de plis

```python
plis = np.arange(2, 20, 1)
accuracy_score = []

for i in plis:
    cv_score = cross_val_score(mlp_meilleur, X_train, y_train, cv=i, scoring="accuracy").mean()
    accuracy_score.append(cv_score)

plt.figure(figsize=(10, 6))
plt.plot(plis, accuracy_score, label='Accuracy en fonction du nombre de plis')
plt.xlabel("Nombre de plis")
plt.ylabel('Score Accuracy')
plt.title("Score Accuracy en fonction du nombre de plis")
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Obtenir l'indice du score d'accuracy maximal
index_meilleur_acc_score = np.argmax(accuracy_score)  # Indice du score d'accuracy maximal
meilleur_nb_plis = plis[index_meilleur_acc_score]  # Nombre de plis correspondant
meilleur_score = accuracy_score[index_meilleur_acc_score]  # Score d'accuracy maximal

print(f"La meilleure configuration est de {meilleur_nb_plis} plis avec un score d'accuracy de {meilleur_score:.4f}.")
```

# GridSearch

```python
from sklearn.model_selection import GridSearchCV
```

```python
# Modèle
mlp_meilleur2 = MLPClassifier(hidden_layer_sizes=(30,)*21, random_state=0)

# Grid
param_grid = {
    'activation': ['relu', 'tanh', 'logistic'],
    'max_iter': range(150, 501, 50),  # 150 à 500 avec un pas de 50
    'alpha': [0.0001, 0.001, 0.01],  # Vous pouvez ajuster cette liste
    'learning_rate': ['constant', 'adaptive']
}

grid_search = GridSearchCV(estimator = mlp_meilleur2, param_grid=param_grid, scoring='f1', cv=3, n_jobs=-1)

grid_search.fit(X_train, y_train)

print("Meilleurs paramètres :", grid_search.best_params_)
print("Meilleur score :", grid_search.best_score_)


```

```python
from sklearn.metrics import f1_score, accuracy_score
```

```python
# Modèle
meilleur_modele = grid_search.best_estimator_

# Prédictions
y_pred = meilleur_modele.predict(X_test)

# Calcul des scores
f1_test_score = f1_score(y_test, y_pred)
accuracy_test_score = accuracy_score(y_test, y_pred)

# Affichage des résultats
print(f"Score F1 sur l'ensemble de test : {f1_test_score:.4f}")
print(f"Accuracy sur l'ensemble de test : {accuracy_test_score:.4f}")
```
